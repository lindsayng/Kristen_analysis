```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from datetime import datetime
```

```python
initial = pd.read_csv('Patch-seq log - Mouse PatchSeq.csv')
ps = initial[['Date', 'Slice (Lims ID?)', 'extraction pressure applied (mbar)', 'Post patch?', 'Post patch pipette R']]
ps.dropna(subset = ['extraction pressure applied (mbar)'], inplace = True)
ps['extraction pressure applied (mbar)'] = ps['extraction pressure applied (mbar)'].astype(float)
ps['Date'] = ps['Date'].astype(int)
#ps = ps[ps['Date'] > 171001]
ps['Date'] = pd.to_datetime(ps['Date'], format = '%y%m%d')
ps.head()
```

```python
def nucleated(x):
    nucleus = x[(x["Post patch?"] == "Nucleated") | (((x["Post patch?"] == "nucleus_visible") | (x["Post patch?"] == "nucleus_present")) & (x["Post patch pipette R"] >= 500))]
    return nucleus

def partial_nucleated(y):
    partial = y[(y["Post patch?"] == 'Partial-Nucleus') | (((y['Post patch?'] == 'nucleus_present') | (y['Post patch?'] == 'nucleus_visible')) & (y["Post patch pipette R"] <= 499))]
    return partial

def outside_out(z):
    outside = z[(z["Post patch?"] == 'Outside-Out') | (((z['Post patch?'] == 'nucleus_absent') | (z['Post patch?'] == 'no_nuclues_visible')) & (z['Post patch pipette R'] >= 500))]
    return outside

def no_seal(w): 
    no = w[(w["Post patch?"] == 'No-Seal') | (((w['Post patch?'] == 'nucleus_absent') | (w['Post patch?'] == 'no_nucleus_visible')) & (w['Post patch pipette R'] <= 499))]
    return no
    
def entire_cell(v):
    entire = v[(v["Post patch?"] == 'Entire-Cell') | (v['Post patch?'] == 'entire_cell')]
    return entire
```

```python
nuc = nucleated(ps)
nuc["post patch class"] = "a. Nucleated"
part_nuc = partial_nucleated(ps)
part_nuc["post patch class"] = "b. Partial nucleated"
out_out = outside_out(ps)
out_out["post patch class"] = "c. Outside out"
noSeal = no_seal(ps)
noSeal["post patch class"] = "d. No seal"
entirecell = entire_cell(ps)
entirecell['post patch class'] = "e. Entire cell"
```

```python
frames = [nuc, part_nuc, out_out, noSeal]
ps = pd.concat(frames)
ps.head()
```

```python
ps = ps.drop('Date', 1)
ps = ps.drop('Post patch?', 1)
ps = ps.drop('Post patch pipette R', 1)
ps['Cre_line'] = ps['Slice (Lims ID?)'].str[:-18]
ps.head()
```

```python
ps = ps.drop('Slice (Lims ID?)', 1)
ps.head()
```

```python
def low_press(x):
    lowp = x[(x['extraction pressure applied (mbar)'] < 0) & (x['extraction pressure applied (mbar)'] > -25)]
    return lowp

def med_press(y):
    medp = y[(y['extraction pressure applied (mbar)'] < -25) & (y['extraction pressure applied (mbar)'] > -50)]
    return medp

def high_press(z):
    highp = z[(z['extraction pressure applied (mbar)'] < -50)]
    return highp
```

```python
low = low_press(ps)
low['press_cat'] = '0-25'
med = med_press(ps)
med['press_cat'] = '25-50'
high = high_press(ps)
high['press_cat'] = '50+'
```

```python
frames = [low, med, high]
ps = pd.concat(frames)
ps.tail()
```

```python
ps = ps.drop('extraction pressure applied (mbar)', 1)
ps.tail()
```

```python
Sst = ps[ps['Cre_line'] == 'Sst-IRES-Cre']
Sst.tail()
```

```python
Sst1 = Sst.groupby(['press_cat', 'post patch class']).size().reset_index(name = 'counts')
Sst1
```

```python
Sst2 = Sst1.set_index(['press_cat', 'post patch class']).unstack(level = -1).fillna(0)
Sst2
```

```python
Sst_percent = Sst2.div(Sst2.sum(1), axis = 0)*100
Sst_percent
```

```python
Sst_totals = list(Sst2.sum(1))
Sst_totals = [int(x) for x in Sst_totals]
Sst_totals
```

```python
Rbp4 = ps[ps['Cre_line'] == 'Rbp4-Cre_KL100']
Rbp4.tail()
```

```python
Rbp41 = Rbp4.groupby(['press_cat', 'post patch class']).size().reset_index(name = 'counts')
Rbp41
```

```python
Rbp42 = Rbp41.set_index(['press_cat', 'post patch class']).unstack(level = -1).fillna(0)
Rbp42
```

```python
Rbp4_percent = Rbp42.div(Rbp42.sum(1), axis = 0)*100
Rbp4_percent
```

```python
Rbp4_totals = list(Rbp42.sum(1))
Rbp4_totals = [int(x) for x in Rbp4_totals]
Rbp4_totals
```

```python
ax1 = Sst_percent.plot(kind = 'bar', stacked = True, ylim = (0, 125), yticks = [0, 50, 100], width = 0.25, position = 0, edgecolor = 'white')
Rbp4_percent.plot(ax = ax1, kind = 'bar', stacked = True, rot = 0, width = 0.25, position = 1, edgecolor = 'white')

sns.despine()
ax1.spines['left'].set_bounds(0, 100)
ax1.spines['bottom'].set_bounds(-0.5, 2.5)

ax1.set_xlabel('Pressure (mbar)')
ax1.set_ylabel('% of Cells')
ax1.set_title('Post-Patch Outcome vs Extraction Pressure', loc = 'center', fontsize = 16)
ax1.legend(['Nucleated', 'Partial nucleated', 'Outside out', 'No seal'], loc = 'upper center', frameon = False, ncol = 4)


ax1.text(2.3, 2, 'left =\nRbp4\n\nright =\nSst')
x = [0, 1, 2]
for i, n in zip(x, Sst_totals):
    ax1.text(i+0.12, 102, n, ha = 'center', va = 'bottom', fontsize = 8)
for i, n in zip(x, Rbp4_totals):
    ax1.text(i-0.12, 102, n, ha = 'center', va = 'bottom', fontsize = 8)
    
plt.savefig('Postpatch vs Extraction Pressure')
```
